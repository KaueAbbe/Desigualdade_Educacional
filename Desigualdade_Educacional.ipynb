## 1. A média do planeta

```python
anos = [x for x in range(2010, 2022)]
anos
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as px
```

```python
def graf_linha(x, y, titulo, xlabel, ylabel):
  fig = px.line(x = x, y =y )

  fig.update_xaxes(title_text= xlabel, row=1, col=1, showgrid = False,showline=True)
  fig.update_yaxes(title_text=ylabel, row=1, col=1,showgrid = False,showline=True)
  fig.update_xaxes(showline=True, linewidth=2, linecolor='black')
  fig.update_yaxes(showline=True, linewidth=2, linecolor='black')

  fig.update_layout(showlegend=False, plot_bgcolor="white",title_text=titulo)



  fig.show()

graf_linha(anos, media_mundo[1:].values, 'Desigualdade Média do Mundo', 'Anos','Desigualdade')
```

## Desvio Padrão

```python
std_mundo = dados.describe().loc['std']

graf_linha(anos, std_mundo[1:].values, 'Desvio Padrão Desigualdade do Mundo', 'Anos','Desvio Padrão')
```

```python
valoresY= dados.describe().loc['min']

graf_linha(anos, valoresY[1:].values, 'Mínimos de Desigualdade do Mundo', 'Anos','Minimo')
```

## Regiões com Maiores Números de Desigualdade

```python
dados.head()
```

```python
dados_por_regioes = dados.copy()
```

```python
dados_por_regioes[~dados_por_regioes.isna()['Inequality in Education (2021)']].info()
```

```python
dados_por_regioes = dados_por_regioes[~dados_por_regioes.isna()['Inequality in Education (2021)']]
dados_por_regioes
```

```python
group_regioes = dados_por_regioes.groupby('UNDP Developing Regions').apply('mean')
group_regioes
```

```python
regioes = group_regioes.sort_values(by = 'Inequality in Education (2021)').index
regioes
```

```python
regioes_2021 = group_regioes.sort_values(by = 'Inequality in Education (2021)')['Inequality in Education (2021)'].values
regioes_2021
```

```python
def graf_bar(x, y, titulo, xlabel, ylabel):
  fig = px.bar(x = x, y =y )

  fig.update_xaxes(title_text= xlabel, row=1, col=1, showgrid = False,showline=True)
  fig.update_yaxes(title_text=ylabel, row=1, col=1,showgrid = False,showline=True)
  fig.update_xaxes(showline=True, linewidth=2, linecolor='black')
  fig.update_yaxes(showline=True, linewidth=2, linecolor='black')

  fig.update_layout(showlegend=False, plot_bgcolor="white",title_text=titulo)



  fig.show()

graf_bar(regioes, regioes_2021,"Desiguldade por região em 2021", "Regiões", "Desiguldade")
```

## Por País maior e menor desigualdade em 2021

```python
dados.head()
```

```python
dados_por_paises = dados_por_regioes.sort_values(by = 'Inequality in Education (2021)').copy()
```

```python
dados_por_paises.tail()
```

```python
paises_top_desiguais = dados_por_paises.tail()['Country'].values
paises_top_desiguais_valores = dados_por_paises.tail()['Inequality in Education (2021)'].values

graf_bar(paises_top_desiguais, paises_top_desiguais_valores,"Paises com Maiores Desigualdades em 2021", "Países", "Desiguldade")
```

```python
dados_por_paises.head()
```

```python
paises_min_desiguais = dados_por_paises.head()['Country'].values
paises_min_desiguais_valores = dados_por_paises.head()['Inequality in Education (2021)'].values

graf_bar(paises_min_desiguais, paises_min_desiguais_valores,"Paises com Maiores Desigualdades em 2021", "Países", "Desiguldade")
```
